```python
from sklearn.utils import resample
def balance_classes(df, label_column):
    # 分离多数类和少数类样本
    df_majority = df[df[label_column] == 0]
    df_minority = df[df[label_column] == 1]
    
    # 计算需要增加的少数类样本数量
    n_majority = df_majority.shape[0]
    n_minority = df_minority.shape[0]
    
    # 过采样少数类
    df_minority_upsampled = resample(df_minority, 
                                     replace=True,     # 样本放回，意味着可以选择相同的样本多次
                                     n_samples=n_majority,    # 将少数类样本数量增加到与多数类相同
                                     random_state=123)  # 随机种子保证结果的可重复性

    # 合并多数类和上采样后的少数类样本
    df_upsampled = pd.concat([df_majority, df_minority_upsampled])

    return df_upsampled

X_train= pd.DataFrame(X_train)
combined_df = pd.concat([y_train, X_train], axis=1)  
combined_df.columns = ['ModelChoice_Default_Flag'] + list(X_train.columns)
combined_df = balance_classes(combined_df, 'ModelChoice_Default_Flag')
X_train = combined_df.drop('ModelChoice_Default_Flag', axis=1)
y_train = combined_df[['ModelChoice_Default_Flag']]
num_rows = X_train.shape[0]  
print("DataFrame的行数:", num_rows)
```

LR Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV

def bestCofLR(train_X, train_y, best_params = None):
    # 定义一个 pipeline，其中包含预处理和逻辑回归
    pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='mean')),  # 缺失值处理
        ('scaler', StandardScaler()),  # 数据标准化
        ('pca', PCA()),  # 主成分分析
        ('classifier', LogisticRegression())  # 逻辑回归分类器
    ])
    if best_params:
        # 接受外面传入的最优秀的参数
        param_grid = best_params
    else:
        # 定义参数网格，同时搜索 PCA 的组件数和逻辑回归的 C 值
        param_grid = {
            'pca__n_components': [7, 9, 11, 13, 15, 17],
            'classifier__C': [0.0001, 0.001, 0.01]
        }

    # 创建一个 GridSearchCV 对象，以 pipeline 为基础进行参数搜索
    grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy')

    # 使用 GridSearchCV 对象训练模型
    grid_search.fit(train_X, train_y)

    # 输出最优的参数组合
    best_params = grid_search.best_params_
    print("Best parameters:", best_params)

    # 返回 GridSearchCV 对象、准确率、预测结果和预测概率
    return grid_search

```

```python

import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
def analyze_model_performance(y_real,y_pred):
     # Generate the confusion matrix
    cm = confusion_matrix(y_real, y_pred)
    tp, fn, fp, tn = cm.ravel()
 
    # Calculate performance metrics
    accuracy = accuracy_score(y_real, y_pred)
    precision = precision_score(y_real, y_pred)
    recall = recall_score(y_real, y_pred)
    f1 = f1_score(y_real, y_pred)
    fnr = fn / (fn + tp) # False Negative Rate
    fpr = fp / (tn + fp) # False Positive Rate, FPR

    # Plotting the confusion matrix
    fig, ax = plt.subplots(figsize=(8, 8))
    cax = ax.matshow(cm, cmap=plt.cm.Reds)
    plt.title('Confusion Matrix', fontsize=16)
    fig.colorbar(cax)
    ax.set_xlabel('Predicted', fontsize=14)
    ax.set_ylabel('Actual', fontsize=14)

    # Setting tick positions explicitly
    ax.set_xticks([0, 1])
    ax.set_yticks([0, 1])
    ax.set_xticklabels(['0', '1'], fontsize=12)
    ax.set_yticklabels(['0', '1'], fontsize=12)

    # Adding specific numbers within the matrix
    for (i, j), val in np.ndenumerate(cm):
        ax.text(j, i, f'{val}', ha='center', va='center', color='black', fontsize=30)

    # Displaying performance metrics
    textstr = '\n'.join([
        f'Accuracy: {accuracy:.2f}',
        f'Precision: {precision:.2f}',
        f'Recall: {recall:.2f}',
        f'F1 Score: {f1:.2f}',
        f'False Negative Rate: {fnr:.2f}',
        f'False Positive Rate: {fpr:.2f}',
        f'TP: {tp}, FP: {fp}, TN: {tn}, FN: {fn}'
    ])

    # Positioning the text box in an appropriate place
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    ax.text(0.02, -0.1, textstr, transform=ax.transAxes, fontsize=14,
            verticalalignment='top', bbox=props)

    plt.show()
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
def calculate_auc(y_true, y_score):
    """
    计算AUC的函数
    
    参数：
    - y_true: 真实标签
    - y_score: 预测得分
    """
    auc = roc_auc_score(y_true, y_score)
    return auc

def plot_roc_curve(y_true, y_score):
    """
    绘制ROC曲线的函数
    
    参数：
    - y_true: 真实标签
    - y_score: 预测得分
    """
    fpr, tpr, _ = roc_curve(y_true, y_score)
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc="lower right")
    plt.show()
```

```python
def AA_with_modular_fit(X_test_AA, Financial_only_model, Internal_only_model, AA_only_model, col_Financial_only, col_Internal_only, col_AA_only):
    X_test_AA_Financial_only = X_test_AA[col_Financial_only]
    X_test_AA_Internal_only = X_test_AA[col_Internal_only]
    X_test_AA_AA_only = X_test_AA[col_AA_only]
    
    y_pred_Financial_only = Financial_only_model.predict(X_test_AA_Financial_only)
    y_score_Financial_only = Financial_only_model.predict_proba(X_test_AA_Financial_only)[:, 1]
    y_pred_Internal_only = Internal_only_model.predict(X_test_AA_Internal_only)
    y_score_Internal_only = Internal_only_model.predict_proba(X_test_AA_Internal_only)[:, 1]
    y_pred_AA_only = AA_only_model.predict(X_test_AA_AA_only)
    y_score_AA_only = AA_only_model.predict_proba(X_test_AA_AA_only)[:, 1]
    
    # 平均概率法
    y_score_average = (y_score_Financial_only + y_score_Internal_only + y_score_AA_only) / 3

    # 投票法（基于预测类别）
    y_pred_majority_vote = np.array([
        np.bincount([y_pred_Financial_only[i], y_pred_Internal_only[i], y_pred_AA_only[i]]).argmax()
        for i in range(len(y_pred_Financial_only))
    ])

    # 加权平均法（示例权重，需要根据模型性能调整）
    weights = np.array([0.4, 0.2, 0.4])  # 假设给三个模型的概率预测分配权重
    y_score_weighted_average = (
        y_score_Financial_only * weights[0] +
        y_score_Internal_only * weights[1] +
        y_score_AA_only * weights[2]
    ) / weights.sum()
    
    return y_pred_majority_vote, y_score_weighted_average
```

```python
def BB_with_modular_fit(X_test_BB, Financial_only_model, Internal_only_model, BB_only_model, col_Financial_only, col_Internal_only, col_BB_only):
    X_test_BB_Financial_only = X_test_BB[col_Financial_only]
    X_test_BB_Internal_only = X_test_BB[col_Internal_only]
    X_test_BB_BB_only = X_test_BB[col_BB_only]

    y_pred_Financial_only = Financial_only_model.predict(X_test_BB_Financial_only)
    y_score_Financial_only = Financial_only_model.predict_proba(X_test_BB_Financial_only)[:, 1]
    y_pred_Internal_only = Internal_only_model.predict(X_test_BB_Internal_only)
    y_score_Internal_only = Internal_only_model.predict_proba(X_test_BB_Internal_only)[:, 1]
    y_pred_BB_only = BB_only_model.predict(X_test_BB_BB_only)
    y_score_BB_only = BB_only_model.predict_proba(X_test_BB_BB_only)[:, 1]
    
    # 平均概率法
    y_score_average = (y_score_Financial_only + y_score_Internal_only + y_score_BB_only) / 3

    # 投票法（基于预测类别）
    y_pred_majority_vote = np.array([
        np.bincount([y_pred_Financial_only[i], y_pred_Internal_only[i], y_pred_BB_only[i]]).argmax()
        for i in range(len(y_pred_Financial_only))
    ])

    # 加权平均法（示例权重，需要根据模型性能调整）
    weights = np.array([0.4, 0.2, 0.4])  # 假设给三个模型的概率预测分配权重
    y_score_weighted_average = (
        y_score_Financial_only * weights[0] +
        y_score_Internal_only * weights[1] +
        y_score_BB_only * weights[2]
    ) / weights.sum()
    
    return y_pred_majority_vote, y_score_weighted_average
```

```python
def CC_with_modular_fit(X_test_CC, Financial_only_CC_model, Internal_only_model, col_Financial_for_CC, col_Internal_only):
    X_test_BB_Financial_only = X_test_CC[col_Financial_for_CC]
    X_test_BB_Internal_only = X_test_CC[col_Internal_only]

    y_pred_Financial_only = Financial_only_CC_model.predict(X_test_BB_Financial_only)
    y_score_Financial_only = Financial_only_CC_model.predict_proba(X_test_BB_Financial_only)[:, 1]
    y_pred_Internal_only = Internal_only_model.predict(X_test_BB_Internal_only)
    y_score_Internal_only = Internal_only_model.predict_proba(X_test_BB_Internal_only)[:, 1]
    
    # 平均概率法
    y_score_average = (y_score_Financial_only + y_score_Internal_only ) / 2

    # 投票法（基于预测类别）
    y_pred_majority_vote = np.array([
        np.bincount([y_pred_Financial_only[i], y_pred_Internal_only[i]]).argmax()
        for i in range(len(y_pred_Financial_only))
    ])

    # 加权平均法（示例权重，需要根据模型性能调整）
    weights = np.array([0.5, 0.5])  # 假设给三个模型的概率预测分配权重
    y_score_weighted_average = (
        y_score_Financial_only * weights[0] +
        y_score_Internal_only * weights[1] 
    ) / weights.sum()
    
    return y_pred_majority_vote, y_score_weighted_average
```
